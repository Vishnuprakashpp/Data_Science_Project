```python
df = pd.read_table("http://bit.ly/imdbratings",sep=',')
#df = df.rename(columns= {'star_rating':'StarRating'}) its a tidyious method
#new_col = ['StarRating','Title','Content_Rating','Genre','Duration','ActorsList']
#df.columns=new_col
df.columns = df.columns.str.replace('_','')
df.head()
```

```python
df.drop(['starrating'],axis=1)
df['duration'].sort_values(ascending=False)
df.sort_values('duration',ascending=False)
df.sort_values(['genre','duration'])
```

#### extracting data of duration greater than 175

```python
boolean = []
for i in df.duration:
    if i>175:
        boolean.append(True)
    else:
        boolean.append(False)
df[boolean]        
```
